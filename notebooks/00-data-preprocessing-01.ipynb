## 3. Data Prep

```python
print("=== DATA FRAME SHAPE: ===")
print(df.shape)

print("=== DATA FRAME SCHEMA: ===")
print(df.dtypes)

print("=== DATA FRAME HEAD: ===")
display(df.head(5))

print("=== DATA FRAME SUMMARY: ===")
display(df.describe())
```

```python
summary_list = list()

for column in df.columns:
    unique_count = df[column].nunique()

    if df[column].dtype == "object" or df[column].dtype.name == "category":
        value_counts = df[column].value_counts().reset_index()
        value_counts.columns = [column, "Count"]
        value_counts_summary = value_counts.to_string(index=False)

    else:
        value_counts_summary = "N/A"

    summary_list.append([column, unique_count, value_counts_summary])

summary_df = pd.DataFrame(
    summary_list,
    columns=[
        "Column",
        "Number of Unique Values",
        "Value Counts",
    ],
)

display(summary_df)
```

### 3.1 Data types

* Modified date with be converted to datetime


```python
df["ModifiedDate"] = pd.to_datetime(df["ModifiedDate"], errors="coerce")
```

### 3.2 Nulls
* Middle name and suffix are nulls. Since these columns are not used in the analysis, we can drop them.

```python
df_aux = pd.DataFrame()
df_aux["# nulls"] = df.isna().sum()
df_aux["% nulls"] = df.isna().sum() / df.shape[0]
df_aux.style.format({"# nulls": "{:,}".format, "% nulls": "{:,.0%}".format})
```

```python
df.drop(columns=["MiddleName", "Suffix"], inplace=True)
```

### 3.3 Duplicates

```python
# Check for duplicates when grouping by customer ID, first name, last name, and modified date
duplicated_rows = df[
    df.duplicated(
        subset=[
            "CustomerID",
            "FirstName",
            "LastName",
            "ModifiedDate",
        ],
        keep=False,
    )
]
print(
    f"----------> There are {duplicated_rows.shape[0]} duplicated rows in the dataset based on CustomerID, FirstName, LastName, and ModifiedDate."
)
```

### 3.5 Reshaping

* We will be creating new columns 

```python
# 1. Extract the last characters after the first '\'
df["salesperson_name"] = df["SalesPerson"].str.split("\\").str[-1].str[:-1]
df["salesperson_name"].value_counts()
```

```python
# 2. Separate Modification Date into ModifiedYear and ModifiedMonth
df["ModifiedYear"] = df["ModifiedDate"].dt.year
df["ModifiedMonth"] = df["ModifiedDate"].dt.month
df["ModifiedDay"] = df["ModifiedDate"].dt.day

print(df[["ModifiedDate", "ModifiedYear", "ModifiedMonth", "ModifiedDay"]].head())
```

```python
# 3. Create a Gender column based on Title
df["Gender"] = df["Title"].apply(
    lambda x: "Male"
    if pd.notnull(x) and ("mr" in str(x).lower() or "sr" in str(x).lower())
    else (
        "Female"
        if pd.notnull(x)
        and (
            "mrs" in str(x).lower() or "ms" in str(x).lower() or "sra" in str(x).lower()
        )
        else "Unknown"
    )
)

print(df[["Title", "Gender"]].head())
```
